```python
import networkx as nx
import matplotlib.pyplot as plt

# Parâmetros do modelo Watts-Strogatz
n = 30  # Número de nós
k = 4   # Cada nó será conectado a k vizinhos mais próximos
p = 0.3 # Probabilidade de reconectar arestas

# Gerando o grafo baseado no modelo Watts-Strogatz
ws_graph = nx.watts_strogatz_graph(n, k, p)

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.circular_layout(ws_graph)  # Posicionamento circular para visualizar os clusters
nx.draw(ws_graph, pos, node_size=300, node_color='blue', edge_color='gray', with_labels=True)

# plt.title(f"Modelo Watts-Strogatz (n={n}, k={k}, p={p})")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Extraindo os graus dos nós
degrees = [ws_graph.degree(n) for n in ws_graph.nodes()]

# Plotando a distribuição de grau
plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=np.arange(min(degrees), max(degrees) + 0.5, 0.5), density=True, color='blue', alpha=0.75, rwidth=0.9)
plt.xlabel("Grau")
plt.ylabel("Frequência")

plt.show()
```
