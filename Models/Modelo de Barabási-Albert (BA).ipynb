```python
import networkx as nx
import matplotlib.pyplot as plt

# Parmetros do modelo Barabsi-Albert
n = 100  # Nmero total de ns na rede
m = 2    # Nmero de arestas que cada novo n adiciona

# Gerando o grafo baseado no modelo Barabsi-Albert
ba_graph = nx.barabasi_albert_graph(n, m)

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(ba_graph)  # Posiciona os ns para uma visualizao mais clara
nx.draw(ba_graph, pos, node_size=30, node_color='blue', edge_color='gray', with_labels=False)

# plt.title("Modelo de Barabsi-Albert (n = 100, m = 2)")
plt.show()
```

```python
import numpy as np

# Extraindo os graus dos ns
degrees = [ba_graph.degree(n) for n in ba_graph.nodes()]

# Plotando a distribuio de grau
plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=np.arange(min(degrees), max(degrees) + 1, 1), density=True, color='blue', alpha=0.75, rwidth=0.95)
# plt.title("Distribuio de Grau no Modelo de Barabsi-Albert")
plt.xlabel("Grau")
plt.ylabel("Frequncia")

plt.show()
```
