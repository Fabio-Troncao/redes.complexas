```python
import networkx as nx
import matplotlib.pyplot as plt

# Parâmetros do modelo Barabási-Albert
n = 100  # Número total de nós na rede
m = 2    # Número de arestas que cada novo nó adiciona

# Gerando o grafo baseado no modelo Barabási-Albert
ba_graph = nx.barabasi_albert_graph(n, m)

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(ba_graph)  # Posiciona os nós para uma visualização mais clara
nx.draw(ba_graph, pos, node_size=30, node_color='blue', edge_color='gray', with_labels=False)

# plt.title("Modelo de Barabási-Albert (n = 100, m = 2)")
plt.show()
```

```python
import numpy as np

# Extraindo os graus dos nós
degrees = [ba_graph.degree(n) for n in ba_graph.nodes()]

# Plotando a distribuição de grau
plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=np.arange(min(degrees), max(degrees) + 1, 1), density=True, color='blue', alpha=0.75, rwidth=0.95)
# plt.title("Distribuição de Grau no Modelo de Barabási-Albert")
plt.xlabel("Grau")
plt.ylabel("Frequência")

plt.show()
```
