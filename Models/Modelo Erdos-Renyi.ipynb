```python
import networkx as nx
import matplotlib.pyplot as plt

# Parmetros do modelo Erds-Rnyi
n = 100  # Nmero de ns
p = 0.05  # Probabilidade de conexo entre os ns

# Gerando o grafo baseado no modelo Erds-Rnyi
er_graph = nx.erdos_renyi_graph(n, p)

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(er_graph)  # Posicionamento para melhor visualizao
nx.draw(er_graph, pos, node_size=30, node_color='blue', edge_color='gray', with_labels=False)

plt.title(f"Modelo Erds-Rnyi (n = {n}, p = {p})")
plt.show()

```

```python
import numpy as np

# Extraindo os graus dos ns
degrees = [er_graph.degree(n) for n in er_graph.nodes()]

# Plotando a distribuio de grau
plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=np.arange(min(degrees), max(degrees) + 1, 1), density=True, color='blue', alpha=0.75, rwidth=0.95)
# plt.title("Distribuio de Grau no Modelo Erds-Rnyi")
plt.xlabel("Grau")
plt.ylabel("Frequncia")

plt.show()
```
