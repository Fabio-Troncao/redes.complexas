```python
import networkx as nx
import matplotlib.pyplot as plt

# Parâmetros do modelo Erdős-Rényi
n = 100  # Número de nós
p = 0.05  # Probabilidade de conexão entre os nós

# Gerando o grafo baseado no modelo Erdős-Rényi
er_graph = nx.erdos_renyi_graph(n, p)

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(er_graph)  # Posicionamento para melhor visualização
nx.draw(er_graph, pos, node_size=30, node_color='blue', edge_color='gray', with_labels=False)

plt.title(f"Modelo Erdős-Rényi (n = {n}, p = {p})")
plt.show()

```

```python
import numpy as np

# Extraindo os graus dos nós
degrees = [er_graph.degree(n) for n in er_graph.nodes()]

# Plotando a distribuição de grau
plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=np.arange(min(degrees), max(degrees) + 1, 1), density=True, color='blue', alpha=0.75, rwidth=0.95)
# plt.title("Distribuição de Grau no Modelo Erdős-Rényi")
plt.xlabel("Grau")
plt.ylabel("Frequência")

plt.show()
```
